# 재귀호출이란?
재귀 호출은 함수가 자기 자신을 다시 호출하는 것을 뜻함   


# 1. 문제 설명
1부터 n 까지의 연속한 정수의 곱을 구하는 알고리즘을 만들어 보세요

> **1-1 팩토리얼이란**    
> * 팩토리얼은 숫자 뒤에 느낌표를 붙여 1부터 n까지 연속한 숫자를 차례로 곱한 값.         
> * "계승"이라고도 한다   
> $ 1! = 1 $   
> $ 3! = 1 * 2 * 3 = 6 $   
> $ 5! = 1 * 2 * 3 * 4 * 5 = 120 $    
> $ n! = 1 * 2 * 3 * 4 ...* (n - 1) * n $   
> 단, 0! 은 1이라고 약속합니다.

## 2. 알고리즘

```python
def fact(n):
    f = 1
    for i in range(1, n + 1):
        f = f * i
        
    return f
```

```python
print(fact(1))
print(fact(5))
print(fact(10))
```

## 3. 재귀호출 예제

```python
def hello():
    print("hello")
    
    hello() # hello() 함수 안에서 다시 hello() 호출
```

```python
hello()
```

### 3-1 해설 
"hello" 를 출력한 후 다시 자기 자신을 호출하므로 또 다시 "hello"를 출력하고 다시 출력하고 반복해 "hello" 출력을 영원히 반복
* 이와 같은 현상을 재귀 에러 (Recursion Error) 라 표현한다. 
* 그러나 hello() 함수는 올바른 재귀 호출 프로그램이라 볼 수 없다
* 재귀 호출을 정상적으로 작동하려면 '종료 조건'이 필요 

## 4. 재귀 호출 알고리즘

```python
def fact_rec(n):
    if n <= 1:
        return 1 
    
    return n * fact_rec(n - 1)
```

```python
print(fact_rec(1))
print(fact_rec(5))
print(fact_rec(10))
```

### 4-1 해설
> 반복되는 fact(n) 속에서 if n <= 1 n이 1에 도달하면 return 1을 반환하므료 종료조건이 된다.     
> n!을 구하려면 곱셉이 (n-1) 번 필요하다. 그러므로 계산복잡도는 $O(n)$

### 5-1 1부터 n까지의 합 구하기를 재귀 호출로 만들어 보세요.

```python
def sum_rec(n):
    if n <= 1:
        return 1
    
    return n + sum_rec(n-1)
```

```python
print(sum_rec(10))
print(sum_rec(100))
```

### 5-2 숫자 n개 중에서 최댓값 찾기를 재귀 호출로 만들어 보세요.

```python
def max_rec(a, n):
    
    if n == 1: 
        return a[0]
    
    max_num = max_rec(a, n-1)
    
    if max_num > a[n-1]:
        return max_num
    
    else:
        return a[n-1]
```

```python
v = [17, 92, 18, 33, 58, 7, 33, 42]

print(max_rec(v, len(v)))
```

**5-2는 1시간 동안 못 풀고 정답 확인하였음**

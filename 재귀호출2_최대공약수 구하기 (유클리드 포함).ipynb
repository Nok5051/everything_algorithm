# 1. 문제 설명 
두 자연수 a 와 b의 최대공약수를 구하는 알고리즘을 만들어 보세요.

> **최대공약수란 ?**   
> 두 개 이상의 정수의 공통 약수 중에서 가장 큰 값을 의미


## 2. 알고리즘

```python
def gcd(a, b):
    i = min(a, b) 
    
    while True:
        
        if a % i == 0 and b % i == 0:
            return i
        
        i = i - 1        
```

```python
print(gcd(1, 6))
print(gcd(3, 6))
print(gcd(60, 24))
print(gcd(81, 27))
```

### 2-1 해설 
> * gcd(60, 24) 가 12가 나온 이유를 보면 문제가 이해가 됨   
> * if문에서 나머지의 값이 and 조건이기 떄문에 나머지가 동시에 0이 되지 않으면 i 값을 -1 해서 0이 나올때까지 반복문을 실행함

## 3. 유클리드 알고리즘
수학자 유클리드 방식을 이용해 최대공약수를 구하기 (재귀호출)

```python
def gcd_rec(a, b):
    if b == 0:
        return a 
    
    return gcd(b, a % b)
```

```python
print(gcd_rec(1, 5))
print(gcd_rec(3, 6))
print(gcd_rec(60, 24))
print(gcd_rec(81, 27))
```

### 3-1 해설
> 1. gcd_rec(3, 6) 으로 해설해 보자면 맨 처음 수행하는 return 값이 gcd_rec(b, a % b) 이기 때문에 gcd_rec(6, 3 % 6)= gcd_rec(6, 3) 값이 출력     

> 2. 이후 다시 재귀호춣로 gcd_rec(6, 3) 수행되고  gcd_rec(3, 6 % 3)= gcd_rec(3, 0)으로 수행

> 3. 이후 재귀호출로 gcd_rec(3, 0)이 수행되고 조건 b == 0 충족함으로 결과값으로 3을 반출
